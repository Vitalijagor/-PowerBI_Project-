```python
Salys = [ 'Poland', 'Lithuania', 'Latvia', 'Estonia']
Reitingas = [40, 42, 53, 55]
plt.bar(Salys, Reitingas)

plt.title('reitingas pagal salys')
plt.xlabel('Salys')
plt.ylabel('reitingas')

plt.show()
```

```python
Salys = [ 'Poland', 'Lithuania', 'Latvia', 'Estonia']
Sveiko_gyvenimo_trukme = [0.884, 0.818, 0.812, 0.874]
plt.bar(Salys, Sveiko_gyvenimo_trukme)

plt.title('Sveiko gyvenimo trukme')
plt.xlabel('Salys')
plt.ylabel('Sveiko gyvenimo trukme')

plt.show()
```

```python
data = {
    'Salys': ['Poland', 'Lithuania', 'Latvia', 'Estonia'],
    'Korupcijos_suvokimas': [0.05, 0.042, 0.064, 0.161]
}

df = pd.DataFrame(data)

# Surūšiuojame šalis pagal korupcijos lygį mažėjimo tvarka
df = df.sort_values(by='Korupcijos_suvokimas', ascending=True)


print(df)
```
