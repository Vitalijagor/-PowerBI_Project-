```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('weather.xls', skiprows=6)
data = data[data['T'].notna()]
data['date'] = pd.to_datetime(data['Local time in Vilnius (airport)'], dayfirst=True)
```

```python
data['dayofyear'] = data['date'].dt.dayofyear
import numpy as np
# Go to radians, create a new variable with cos
data['cos_dayofyear'] = np.cos((data['dayofyear'] - 1) / 366 * 2 * np.pi)
```

```python
data_train = data[data['date'] < '2024-01-01']
data_test = data[data['date'] >= '2024-01-01']
x_train = pd.DataFrame()
x_train['cos_dayofyear'] = data_train['cos_dayofyear']
x_test = pd.DataFrame()
x_test['cos_dayofyear'] = data_test['cos_dayofyear']
y_train = data_train['T']
y_test = data_test['T']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Prediction for training data
pred_train = model.predict(x_train)
# Prediction for data that the model has not yet seen
pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print('Sum of errors on training data =', mean_squared_error(y_train, pred_train))
print('Sum of errors on test data =', mean_squared_error(y_test, pred_test))
```

```python
plt.figure(figsize=(20, 5))
plt.scatter(data_train['dayofyear'], y_train, label='Train y(x)')
plt.scatter(data_test['dayofyear'], y_test, label='Test y(x)')
plt.scatter(data_test['dayofyear'], pred_test, label='Test predict', color='red')
plt.legend()
plt.show()
```
